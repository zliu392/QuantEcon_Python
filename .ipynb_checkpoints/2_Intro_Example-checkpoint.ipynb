### (3) Loop

### (4) User Defined Functions

```python
def generate_data(n, generator_type):
    epsilon_values = []
    for i in range(100):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        epsilon_values.append(e)
    return epsilon_values

data = generate_data(100, 'U')
plt.plot(data, 'b-')
plt.show()
```

### (6) Exercises

#### Q1. Factorial

```python
## 1
def factorial(n):
    fact = []
    if type(n) == int:
        fact = n
        while n > 1:
            fact = fact * (n-1)
            n = n - 1
    else:
        fact = np.NaN
        print('The input is not an integer.')
    return fact

factorial(4)
```

#### Q2. Binomial

```python
## 2
from numpy.random import uniform
def binomial_rv(n, p):
    bino = 0
    for i in range(n):
        if uniform(size=1) < p:
            bino += 1
    return bino

binomial_rv(10,0.4)    
```

#### Q3. Compute $\pi$

```python
## 3
import numpy as np
def compute_PI(n):
    times = 0
    for i in range(n):
        a = np.random.uniform(size = 1)
        b = np.random.uniform(size = 1)
        if np.sqrt((a-0.5)**2 + (b-0.5)**2) < 0.5:
            times += 1
    PI = times/n/0.25
    return PI

compute_PI(10000)
```

#### Q5. AR(1)

```python
## 5
import numpy as np
import matplotlib.pyplot as plt
T = 200
alpha = 0.9
current_x = 0
x_values = []
for i in range(T+1):
    x_values.append(current_x)
    current_x = alpha * current_x + np.random.normal()
plt.plot(x_values, 'b-')
plt.show()
len(x_values)
```

Tips: 用current_x生成变量, x_values的append方法来记录

#### Q6. White Noise

```python
## 6
import numpy as np
import matplotlib.pyplot as plt

x = [np.random.randn() for i in range(100)]
# 或者 x = np.random.normal(size = 100)
plt.plot(x, 'b-', label="white noise")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

alphas = [0, 0.8, 0.98]
T = 200
for alpha in alphas:
    current_x = 0
    x_value = []
    for i in range(T+1):
        x_value.append(current_x)
        current_x = alpha * current_x + np.random.randn()
    plt.plot(x_value, label = 'alpha = ' + str(alpha))
plt.legend()
plt.show()
```
